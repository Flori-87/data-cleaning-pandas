```python
index_location= list(Location_for_country2.index)
index_area= list(Area_for_country2.index)
index_coincidencia = []
for e in index_location:
    if e in index_area:
        index_coincidencia.append(e)
print(index_coincidencia)
```

```python
Location_coincide_area = pd.DataFrame(Location_for_country2.loc[index_coincidencia])
display(Location_coincide_area)
Area_coincide_location = pd.DataFrame(Area_for_country2.loc[index_coincidencia])
Area_coincide_location
```

```python
lista = ["Northern Atlantic","Indian","Northern Atlantic","Indian","Southern Atlantic"]
Oceans = pd.DataFrame(lista,index=index_coincidencia,columns=["Ocean"])
Oceans
```

```python
Area_for_country3 = Area_for_country2.drop(index=index_coincidencia)
display(Area_for_country3)
Location_for_country3 = Location_for_country2.drop(index=index_coincidencia)
display(Location_for_country3)
```

```python
# en un unico dataframe las areas y locations que faltan en countries y las comunes que he llamado ocean
a = list(Area_for_country3.index)+list(Location_for_country3.index)+list(Oceans.index)
b = Area_for_country3.values
b2 = Location_for_country3.values
b3 = Oceans.values
c =[i for sublist in b for i in sublist]
d = [i for sublist in b2 for i in sublist]
e = [i for sublist in b3 for i in sublist]
f = c+d+e

def dictionary(x,y):
    dicti={}
    for i in range(len(x)):
        dicti[x[i]]=y[i]
    return pd.DataFrame([dicti[key] for key in dicti.keys()], index= dicti.keys(),columns=["Area"])

falta_en_country =dictionary(a,f)
falta_en_country = falta_en_country.sort_index()
falta_en_country
```

```python
def nodata(value):
    if value == "NoData":
        return value
    else:
        if re.search(r'NoData.*',value):
            a = value.split("NoData")
            b = a[1]
            return b
        elif re.search(r'.*NoData',value):
            a = value.split("NoData")
            b = a[0]
            return b

print(nodata("NoDataUSA"))
print(nodata("USANoData"))
```

```python
concat["Suma2"] = concat["suma"].apply(nodata)
```

A priori podría parecer que la carne de hombre es más apetecible para los tiburones. Sin embargo, vamos a comparar sex con date

```python
attacks1["hombre"] = np.where(attacks1["Sex_2"].isin(["M"]), 1, 0)
attacks1["mujer"] = np.where(attacks1["Sex_2"].isin(["F"]), 1, 0)
attacks1["Fatal"] = np.where(attacks1["Fatal (Y/N)"].isin(["Y"]), 1, 0)
attacks1["NonFatal"] = np.where(attacks1["Fatal (Y/N)"].isin(["N"]), 1, 0)
attacks1
```

```python
attacks1["Date"] = attacks1["Date"].fillna("NoData")
def date_fn(value):
    if (len(value)==4) | (value=="NoData"):
        return value
    else:
        pattern = r'\d{4}\-\d{4}.*'
        if re.match(pattern, value):
            value = value.split("-")
            value1 = value[0]
            value2 = value[1]
            value2 = re.search(r"^\d{4}", value2)
            value2 = value2.group()
            value = [int(value1),int(value2)]
            value_mean = int(sum(value)/len(value))
            return str(value_mean)
        else: 
            value = (re.search(r"^\d{4}", value)).group()
            return value


attacks1["Date_mod"] = attacks1["Date"].apply(date_fn)
attacks1["Date_mod"]
def date_int(value):
    if value == "NoData":
        return 0
    else:
        value = int(value)
        if value>2019:
            value=0
            return value
        else:
            return value

attacks1["Date_mod2"] = attacks1["Date_mod"].apply(date_int)
attacks1.loc[attacks1["Date_mod2"] == 0,"Date_mod2"] = np.nan
```

```python
attacks1 = attacks1.drop(columns=["Fatal (Y/N)","Sex","Sex_2"])
```

```python
attacks1 = attacks1.drop(columns=["Date","Date_mod"])
attacks1                         
```

```python
attacks1["Fatal mujer"] = ((attacks1["Fatal"]==attacks1["mujer"]) & (attacks1["Fatal"]==1))
attacks1["Fatal hombre"] = ((attacks1["Fatal"]==attacks1["hombre"]) & (attacks1["Fatal"]==1))
```
